```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
cityDataDF = pd.read_csv(city_data_to_load)
rideDataDF = pd.read_csv(ride_data_to_load)
#cityDataDF

# Combine the data into a single dataset 
#mergeDF = pd.merge(cityDataDF, rideDataDF)
mergeDF = pd.merge(rideDataDF, cityDataDF, on='city')
                           #left_on=None,
                           #right_on=None, left_index=False, right_index=False, 
                           #sort=False, suffixes=('_x', '_y'), copy=False, indicator=False, 
                           #validate=None)
# Display the data table for preview
mergeDF
```

```python
'''urbanCity = mergeDF[mergeDF["type"]=="Urban"].groupby("city").mean() #creates an ??array? of the city, fare, ride_id, and driver_count
urbanCity = pd.DataFrame(urbanCity) #creates a data frame
averageFareUrban = urbanCity["fare"] #calculates the average fare for each city in the csv file
#urbanCityRides = pd.DataFrame(mergeDF[mergeDF["type"] == "Urban"])
urbanCityRides = urbanCity["city"].value_counts()
urbanCityRides
'''
urban = mergeDF[mergeDF["type"]=="Urban"] #creates an ??array? of the city, fare, ride_id, and driver_count
urbanCity = urban.groupby("city").mean()
#urbanCity = pd.DataFrame(urbanCity) #creates a data frame
averageFareUrban = urbanCity["fare"] #calculates the average fare for each city in the csv file
#urbanCityRides = pd.DataFrame(mergeDF[mergeDF["type"] == "Urban"])
urbanCityRides = urban["city"].value_counts()
urbanCityRides
#################################################################################################################
suburban = mergeDF[mergeDF["type"]=="Suburban"] #creates an ??array? of the city, fare, ride_id, and driver_count
suburbanCity = suburban.groupby("city").mean()
#suburbanCity = pd.DataFrame(suburbanCity) #creates a data frame
averageFareSuburban = suburbanCity["fare"] #calculates the average fare for each city in the csv file
#suburbanCityRides = pd.DataFrame(mergeDF[mergeDF["type"] == "Urban"])
suburbanCityRides = suburban["city"].value_counts()
suburbanCityRides
#################################################################################################################
rural = mergeDF[mergeDF["type"]=="Rural"] #creates an ??array? of the city, fare, ride_id, and driver_count
ruralCity = rural.groupby("city").mean()
#ruralCity = pd.DataFrame(ruralCity) #creates a data frame
averageFareRural = ruralCity["fare"] #calculates the average fare for each city in the csv file
#ruralCityRides = pd.DataFrame(mergeDF[mergeDF["type"] == "Urban"])
ruralCityRides = rural["city"].value_counts()
#ruralCityDrivers = 
#################################################################################################################
driversPerCity = mergeDF.groupby("city").mean()
driversPerCity = driversPerCity["driver_count"]
driversPerCity

#################################################################################################################
urbanPlot = plt.scatter(urbanCityRides, averageFareUrban, s=driversPerCity*7, color="Gold", alpha=1, edgecolor="gray", label="Urban", linewidth=1)
suburbanPlot = plt.scatter(suburbanCityRides, averageFareSuburban, s=driversPerCity*7, color="Coral", alpha=1, edgecolor="gray", label="Suburban", linewidth=1)
ruralPlot = plt.scatter(ruralCityRides, averageFareRural, s=driversPerCity*7, color="Skyblue", alpha=1, edgecolor="gray", label="Rural", linewidth=1)

#################################################################################################################
plt.title("Pyber Ride Sharing Data (2018)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.text(42, 42, s="Note: Circle size correlates with driver count per city.")
plt.grid()
#################################################################################################################
plt.legend( handles=[urbanPlot, suburbanPlot, ruralPlot], loc="best")
#################################################################################################################
plt.savefig("PyberRideSharingData_BubblePlot.png")
plt.tight_layout()
```

## Total Fares by City Type

```python
# Calculate Type Percents
totalUrbanFares = urban["fare"].sum()
totalSuburbanFares = suburban["fare"].sum()
totalRuralFares = rural["fare"].sum()

# Build Pie Chart
labels = ["Urban", "Suburban", "Rural"]
colors = ["Gold", "Coral", "Skyblue"]
explode = (0.08, 0.03, 0.03)
totalFaresByType = [totalUrbanFares, totalSuburbanFares, totalRuralFares]
plt.pie(totalFaresByType, labels=labels, colors=colors, explode=explode, shadow=True, autopct="%1.1f%%", startangle=270)
#plt.axis("equal")
plt.title("% of Total Fares by City Type")
plt.show()
# Save Figure
plt.savefig("totalFaresByCityType_PieChart.png")
plt.tight_layout()
```
